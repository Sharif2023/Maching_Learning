```python
import numpy as np
import matplotlib.pyplot as plt
```

### Create some sample data (X and y)

```python
# y = 2x + 1 + noise

np.random.seed(42)  # for reproducibility
X = 2 * np.random.rand(100, 1)   # random values between 0 and 2
y = 4 + 3 * X + np.random.randn(100, 1)  # linear relation with some noise

plt.scatter(X, y, color="blue")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Data Points")
plt.show()
```

### Add bias term (X0 = 1) for matrix calculation

```python
X_b = np.c_[np.ones((100, 1)), X]   # add column of 1s
X_b[:5]  # show first 5 rows
```

### Calculate the best theta (parameters) using Normal Equation

```python
# Formula: theta = (X^T X)^(-1) X^T y

theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

### Make predictions

```python
X_new = np.array([[0], [2]])  # test values
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add bias
y_predict = X_new_b.dot(theta_best)

y_predict
```

### Plot the regression line

```python
plt.scatter(X, y, color="blue")
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Prediction")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

### Compare with Scikit-Learn LinearRegression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Intercept:", lin_reg.intercept_)
print("Slope:", lin_reg.coef_)

lin_reg.predict([[0], [2]])
```
